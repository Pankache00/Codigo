# CARGA DE LOS DATOS


```python
# Conectamos con la base de datos y en caso de no existir se creará la nueva base de datos
conn = sqlite3.connect("Zoo.db")

# Cargar DataFrames a la base de datos
# Usamos if_exists="replace" para evitar errores si la tabla ya existe

df_zoo.to_sql("Zoologico", conn, if_exists="replace")
df_especies.to_sql("Especie", conn, if_exists="replace")
df_encargados.to_sql("Encargado", conn, if_exists="replace")
df_animales.to_sql("Animal", conn, if_exists="replace")
df_veterinarios.to_sql("Veterinario", conn, if_exists="replace")
df_diagnostico.to_sql("RegistroMedico", conn, if_exists="replace")
df_categoria.to_sql("CategoriaDiagnostico", conn, if_exists="replace")
```

```python
# Consulta de relación entre 3 entidades: Animal, Especie y Zoologico.
# Se utiliza JOIN para unir las tablas mediante claves foráneas,
# obteniendo información combinada con los nombres de columnas originales.

consulta = """
SELECT 
    A.nombre,
    A.sexo,
    A.año_nacimiento,
    E.nombre_comun,
    E.estatus_conservacion,
    Z.nombre_zoologico,
    Z.ciudad
FROM Animal A
JOIN Especie E ON A.especie_id = E.id_especie
JOIN Zoologico Z ON A.zoo_id = Z.id_zoo
"""
resultado = pd.read_sql_query(consulta, conn)
# Una vez realizada la consulta mostramos el resultado de la Query
display(resultado)
```

```python
# Consulta que relaciona diagnóstico, animal, especie, zoológico, veterinario y categoría.
# Muestra información médica completa, incluyendo fecha de atención, diagnóstico, tratamiento y nivel de gravedad.
# Se utilizan JOINs para unir correctamente todas las entidades involucradas.

consulta_diagnostico = """
SELECT 
    R.id_registro,
    R.fecha_atencion,
    R.diagnóstico,
    R.tratamiento,
    A.nombre,
    E.nombre_comun,
    Z.nombre_zoologico,
    V.nombre_especialista,
    C.rango
FROM RegistroMedico R
JOIN Animal A ON R.animal_id = A.id_animal
JOIN Especie E ON A.especie_id = E.id_especie
JOIN Zoologico Z ON A.zoo_id = Z.id_zoo
JOIN Veterinario V ON R.veterinario_id = V.id_veterinario
JOIN CategoriaDiagnostico C ON R.categoria_id = C.id_categoria
"""

resultado = pd.read_sql_query(consulta_diagnostico, conn)


def color_rango(val):
    val = str(val).lower()
    if val == "verde":
        return "background-color: lightgreen"
    elif val == "amarillo":
        return "background-color: yellow"
    elif val == "rojo":
        return "background-color: lightcoral"
    return ""


resultado.style.applymap(color_rango, subset=["rango"])
```
